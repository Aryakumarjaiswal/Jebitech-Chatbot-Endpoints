## DATA PREPROCESSING:


```python
df2.rename(columns={'id': 'property_id'},inplace=True)#Renaming guesty Listing Column:'id'-> 'property_id'
df2.columns
```

```python
#Full outer join on 'property_id'
final_sheet=pd.merge(df1,df2,on='property_id',how='outer')
final_sheet
```

```python
import pandas as pd

# Ensure values are strings and strip spaces
final_sheet["property_id"] = final_sheet["property_id"].astype(str).str.strip()

# Remove completely empty strings
final_sheet = final_sheet[final_sheet["property_id"] != ""]

# Convert to float, forcing errors to NaN
final_sheet["property_id"] = pd.to_numeric(final_sheet["property_id"], errors="coerce")

# Drop NaNs to avoid IntCastingNaNError
final_sheet = final_sheet.dropna(subset=["property_id"])

# Convert to integer safely
final_sheet["property_id"] = final_sheet["property_id"].astype(int)
```

## CHUNCK CREATION:

```python
import chromadb
```

```python
client = chromadb.PersistentClient(path='CHUNCKS')
```

```python

def chunk_text(text, chunk_size=120):
    words = text.split()
    chunks = []
    
    for i in range(0, len(words), chunk_size):
        chunks.append(' '.join(words[i:i+chunk_size]))
    
    return chunks


for idx, row in final_sheet.iterrows():
    doc_id = str(row['property_id'])
    combined_info = row['summary']
    
    # Create a collection for each ID
    collection = client.get_or_create_collection(
        name=f"collection_{doc_id}",
        metadata={"id": doc_id},
    )
    
    
    chunks = chunk_text(combined_info)
    for i,chunk in enumerate(chunks):
        collection.add(
            documents=[chunk],
            metadatas=[{"chunk_index": i}],
            ids=[f"{doc_id}_{i}"]
        )

print("Data successfully stored in ChromaDB.")
```

```python
collection_id = "collection_910"
query_text = "Please tell Entry key of the property"
collection = client.get_collection(collection_id)
results = collection.query(
    query_texts=[query_text],
    n_results=3  # Retrieve top 3 matches
)

print("Query Results:")
print(results)
```
